```python
prefectures_data = []

url_start = 'https://ja.wikipedia.org/wiki/'
url_tail = '日本の郷土料理一覧'

url = url_start + url_tail
print(url)

result = requests.get(url)
c = result.content
    
# HTMLをもとに、オブジェクトを作る
soup = BeautifulSoup(c, "lxml")

prefectures = soup.find_all('span',{'class':'mw-headline'})

for item in range(len(prefectures)):
    prefectures_text = prefectures[item].find(text=True)
    #print(prefectures_text)
    if re.compile('地方').search(prefectures_text):
        #print(re.compile('地方').search(prefectures_text))
        prefectures_data.append(prefectures_text)
    
# tableを見つけます。
tables = soup.find_all('table',{'class':'wikitable'})

#print(tables)
```

```python
# データを格納するためのリストです。
data = []
data_table = []
data_prefectures = []
data_1 = []
data_2 = []
data_3 = []
data_4 = []
data_5 = []
data_6 = []
data_7 = []
```

```python
for item in range(len(prefectures)):
    # テーブルから行をすべて探し出します。
    rows = tables[item].find_all('tr')

    # 行から、それぞれのcellを取り出して画面に表示しつつ、dataに格納します。
    for tr in rows:
        #print(prefectures[item])
        data_prefectures.append(prefectures[item])
        cols = tr.find_all('td')
        # textを探し出します。
        i=1
        for td in cols:
            
            #print(i)
            if (i == 1):
                data_1_csv = []
                #print("i=1")
                text = td.find(text=True) 
                #print(text)
                data_1.append(text)
                data_1_csv.append(text)
                csvWriter.writerow(data_1_csv)          #1行書き込み
                i += 1
                #print(i)
            elif (i == 2):
                #print("i=2")
                text = td.find(text=True) 
                #print(text)
                data_2.append(text)
                i += 1
            elif i == 3:
                #print("i=3")
                text = td.find(text=True) 
                #print(text)
                data_3.append(text)
                i += 1
            elif i == 4:
                #print("i=4")
                text = td.find(text=True) 
                #print(text)
                data_4.append(text)
                i += 1
            elif i == 5:
                #print("i=5")
                text = td.find(text=True) 
                #print(text)
                data_5.append(text)
                i += 1
            elif i == 6:
                #print("i=6")
                text = td.find(text=True) 
                #print(text)
                data_6.append(text)
                i += 1
            elif i == 7:
                #print("i=7")
                text = td.find(text=True) 
                #print(text)
                data_7.append(text)
                i += 1
        #data_prefectures.append(prefectures[item])
        #data_table.append(data)
```
